```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("train.csv")
df.sample(5)
```

```python
df['date']=pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df.shape
```

```python
df["year"]=df["date"].dt.year
```

```python
df["month"]=df["date"].dt.month
df["month_name"]=df["date"].dt.month_name()
df["day"]=df["date"].dt.day
df["day_name"]=df["date"].dt.day_name()
```

```python
df["date"].dt.day_name()
```

```python
df.head(10)
```

```python
start_date=np.min(df["date"])
end_date=np.min(df['date'])
```

```python
# At first we work with store 1 and item 1
```

```python
# selecting store 1
store1=df["store"]==1
```

```python
# selecting item 1
item1=df["item"]==1
```

```python
# selecting store 1 and item 1
df_store1_item1=df[store1 & item1]
```

```python
df_store1_item1.head()
```

```python
df_store1_item1_2013=df_store1_item1[df_store1_item1["year"]==2013]
df_store1_item1_2013.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,8))
sns.barplot(data=df_store1_item1_2013,x="month_name",y="sales")
plt.show()
```

```python
year=df_store1_item1["year"].unique()
```

```python
plt.figure(figsize=(16,25))
i=1
for y in year:
    plt.subplot(len(year),1,i)
    i=i+1
    current_df=df_store1_item1[df_store1_item1["year"]==y]
    sns.barplot(data=current_df,x="month_name",y="sales")
    plt.xlabel(y)
plt.show()
```

# <font color="red">Insight: </font> By analyzing the graph, we observed that sales of item 1 at store 1 are :- <br> 1. Consistently high in July and slightly lower in June each year. <br> 2. Mostly low sale in january month every year.
